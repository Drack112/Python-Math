# Proporção

## Conceito

Proporção é um conceito de Matemática Básica muito útil na própria Matemática e em outras áreas de conhecimento, como a Física e a Química, além da busca de medidas proporcionais na área da Arte. Ao falar de proporção, estamos falando de comparação, então, entendemos como proporção a classificação da comparação entre duas ou mais grandezas.

Mesmo que tenham relação, as grandezas podem relacionar-se de forma não proporcional. A proporção é utilizada para calcular valores desconhecidos por meio de uma razão, que acaba sendo resolvida por equações e regra de três.

## Razão e proporção

A proporção está intimamente ligada com a razão. Para verificação de uma proporcionalidade, ou seja, se as grandezas são proporcionais ou não, devemos fazer uma igualdade de duas razões. Se essas razões forem iguais, os valores serão proporcionais.

# Problema de Proporção com Livros

Dado que 4 livros custam $20, queremos determinar quanto custariam 7 livros.

Podemos configurar a proporção da seguinte maneira:

\[ \frac{4}{20} = \frac{7}{x} \]

Onde:
- \( x \) é o custo de 7 livros.

Usando a técnica de multiplicação cruzada, temos:

\[ 4x = 20 \times 7 \]

Resolva para \( x \) para encontrar o custo de 7 livros.


```python
# Criando uma proporção com multiplicação cruzada

import matplotlib.pyplot as plt
import numpy as np
from sympy import symbols, Eq, solve

x = symbols('x')
a, b, c, d = 4, 20, 7, x

equation = Eq(a*d, b*c)
solution = solve(equation)[0]

print("Equação:")
display(equation)
'''

4    7
-- = --
20   x

4*X=20*7
4x = 149
x = 140/4
x = 35
'''

prices = [b/a * books for books in range(10)]
plt.plot(range(10), prices, '-o', label='Preço dos livros')
plt.axvline(x=c, color='r', linestyle='--', label=f'7 livros = ${solution}')
plt.xlabel('Número de livros')
plt.ylabel('Preço ($)')
plt.title('Preço dos livros em função da quantidade')

plt.legend()
plt.grid(True)
plt.show()
```
